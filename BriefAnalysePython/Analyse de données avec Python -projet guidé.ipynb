```python
nom_colonnes = ['Code','Libelle','Pole emploi 2018']
p_emploi = pd.read_excel('//home//simplon//Téléchargements/pole_emploi.xlsx',
                         names =nom_colonnes,skiprows = 3)


p_emploi.head()
```

```python
data = data.sort_values(by = ['REG'])
p_emploi = p_emploi.sort_values(by = ['Code'])


p_tx_chomage= data.merge(p_emploi,how = 'left' ,left_on = 'REG',right_on = 'Code')
print(p_tx_chomage)
```

```python
p_tx_chomage['quantile']=pd.qcut(p_tx_chomage['Pole emploi 2018'], q=4,labels = ['Q25','Q50','Q75','Q100'])
p_tx_chomage.columns
```

```python

pd.pivot(p_tx_chomage,index = 'LIBZE2010',columns = 'quantile',values= '2020-T1')
```
